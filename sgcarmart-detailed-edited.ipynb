```python
import numpy as np
import pandas as pd

from bs4 import BeautifulSoup as bs
import requests
import re
from selenium import webdriver
```

```python
#driver = webdriver.Chrome('C:\\Users\\65844\\chromedriver.exe')
url = 'https://www.sgcarmart.com/used_cars/info.php?ID=889114&DL=2314'
#driver.get(url)
#elem = driver.find_element_by_xpath("//*")
#source_code = elem.get_attribute("innerHTML")
#sg_car_mart = bs(source_code)
#driver.close()
resp = requests.get(url)
sg_car_mart = bs(resp.text, 'html.parser')
```

```python
print(sg_car_mart.prettify())
```

```python
content = sg_car_mart.findAll('div', {'id':'main_left'})
content2 = content[0].findAll('td',{'valign':'top'})

bulk1 = content[0].findAll('td',{'class':'even_row'})[0].findAll('div',{'class':'row_info'})
bulk2 = content[0].findAll('td',{'class':'even_row'})[1].findAll('div',{'class':'row_info'})
```

```python
vehmake = sg_car_mart.findAll('div',{'id':'toMap'})[0].a.text.split()[0]

price = content[0].find_next('td', {'class':'font_red'}).a.text
vehtype = content[0].find_next('td',{'colspan':'3'}).text
depreciation = content2[0].text.strip().replace(' /yr', '')

mileage = bulk1[0].text.strip().split()[0]
tax = bulk1[1].text.strip().split()[0]
dereg = bulk1[2].text.strip().split()[0]
coe = bulk1[3].text.strip()
eng_cap = bulk1[4].text.strip().split()[0]
curbw = bulk1[5].text.strip().split()[0]

mf_date = bulk2[0].text
transmission = bulk2[1].text
op_mkt_value = bulk2[2].text
arf = bulk2[3].text

print(vehmake, price, vehtype, depreciation, mileage, tax, dereg, coe,
      eng_cap, curbw, mf_date, transmission, op_mkt_value, arf,'\n')
```

```python
combinedlist = []
count = 1
for ids in range(888000,889100,1):
    try:
        url_i = 'https://www.sgcarmart.com/used_cars/info.php?ID={}&DL=1000'.format(ids)
        resp = requests.get(url_i)
        sg_car_mart = bs(resp.text, 'html.parser')
        
        content = sg_car_mart.findAll('div', {'id':'main_left'})
        content2 = content[0].findAll('td',{'valign':'top'})

        bulk1 = content[0].findAll('td',{'class':'even_row'})[0].findAll('div',{'class':'row_info'})
        bulk2 = content[0].findAll('td',{'class':'even_row'})[1].findAll('div',{'class':'row_info'})

        vehmake = sg_car_mart.findAll('div',{'id':'toMap'})[0].a.text.split()[0]
        price = content[0].find_next('td', {'class':'font_red'}).a.text.replace(',','')
        vehtype = content[0].find_next('td',{'colspan':'3'}).text
        depreciation = content2[0].text.strip().split()[0].replace(',','')

        mileage = bulk1[0].text.strip().split()[0].replace(',','')
        tax = bulk1[1].text.strip().split()[0].replace(',','')
        dereg = bulk1[2].text.strip().split()[0].replace(',','')
        coe = bulk1[3].text.strip().replace(',','')
        eng_cap = bulk1[4].text.strip().split()[0].replace(',','')
        curbw = bulk1[5].text.strip().split()[0].replace(',','')

        mf_date = bulk2[0].text
        transmission = bulk2[1].text
        op_mkt_value = bulk2[2].text.replace(',','')
        arf = bulk2[3].text.replace(',','')
        
        print(count)
        count+=1
        print(url_i)
        print(vehmake, price, vehtype, depreciation, mileage, tax, dereg, coe,
              eng_cap, curbw, mf_date, transmission, op_mkt_value, arf,'\n')
        combinedlist.append(vehmake+','+vehtype+','+mileage+','+depreciation+','+tax+','+dereg+','+coe+','+
              eng_cap+','+curbw+','+mf_date+','+transmission+','+op_mkt_value+','+arf+','+price+'\n')
        
    except:
        continue
```

```python
headers = 'Brand,Type,Mil,Dep,Tax,Dereg,Coe,Eng_cap,Curbw,Mf_date,Transmission,Op_mkt_value,Arf,Price\n'
filename = 'usedcarprice.csv'
f = open(filename, 'w')
f.write(headers)

for var in combinedlist:
    f.write(var)

f.close()
```
